```python
def tuble_ec():
  t =(1,2,3)
  return t

a,b,c = tuble_ec()
print(a,b,c)
```

```python
x = 2
def f():
  x = 3
  return x

print(x)  # x = 2 global scope
print(f()) # x == 3 local scope
```

```python
# How can we learn what is built in scope
import builtins
dir(builtins)
```

```python
#nested function
def square():
    """ return square of value """
    def add():
        """ add two local variable """
        x = 2
        y = 3
        z = x + y
        return z
    return add()**2
print(square())   
```

```python
# default arguments
def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5))
# what if we want to change default arguments
print(f(5,4,3))
```

```python
# flexible arguments *args
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)
# flexible arguments **kwargs that is dictionary
def f(**kwargs):
    """ print key and value of dictionary"""
    for key, value in kwargs.items():               # If you do not understand this part turn for loop part and look at dictionary in for loop
        print(key, " ", value)
f(country = 'spain', capital = 'madrid', population = 123456)
```

```python
#  Lambda Fuctions
# user defined function(long way)

def use(x):

  return x*2
print(use(2))

# lambda

use = lambda x : x**2
print(use(2))

tot = lambda x,y,z : x+y+z
print(tot(2,4,5))

```

```python
# Anonymous Functıon
number_list =[5,7,2,9,11]
y = map(lambda x : x**2,number_list)
print(list(y))
```

# LIST COMPREHENSİON¶ and ITERATORS


```python
# iteration example

name = "Onur"

it = iter(name)
print(next(it))
print(next(it))
print(*it)



```

```python
# zip example

liste1=[1,2,3,4]
liste2=[5,6,7,8]
z = zip(liste1,liste2)
un_z = list(z)
print(un_z)
print(list(z))
```

```python
un_zip =zip(*un_z)
un_list1,un_list2 = list(un_zip)
print(un_list1)
print(un_list2)
```

```python
# Example of list comprehension

num1 = [1,2,3,4]
num2 = [i+i for i in num1]
print(num2)
```

```python
# Conditionals on iterable
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
```
